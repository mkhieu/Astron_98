```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
text_df = pd.read_json('AOC_recent_tweets.txt')
text_df
```

```python
def time_in_hours(datetime_column):
    return datetime_column.dt.hour + datetime_column.dt.minute / 60 + datetime_column.dt.second / 3600
text_df['hours'] = time_in_hours(pd.to_datetime(text_df['created_at']))
text_df
```

```python
text_df[['created_at', 'hours', 'full_text']].to_csv('AOC_recent_tweets.csv', index=False)
```

```python
planets = sns.load_dataset('planets')
planets
```

```python
sns.scatterplot(x = 'orbital_period', y = 'mass', data = planets, hue = 'method')
plt.title('Exoplanet Orbital Period vs Exoplanet Mass')
plt.xlabel('Orbital Period')
plt.ylabel('Mass')
plt.xscale('log')
plt.yscale('log')
```

```python
plt.figure(figsize = (12, 6))
sns.countplot(x = 'year', hue = 'method', data = planets, palette = 'Set3', dodge = False)
plt.title('Exoplanet Discoveries by Year')
plt.xlabel('Year')
plt.ylabel('Number of Exoplanets Discovered')
plt.legend(title = 'Discovery Method')
plt.show()
```
