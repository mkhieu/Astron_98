```python
def generatePi(n):
    inlist_x, inlist_y, outlist_x, outlist_y, n_in = [], [], [], [], 0
    for i in range(int(n)):
        x, y = random.random(), random.random()
        dist = math.sqrt((x*x) + (y*y))
        if dist < 1:
            n_in += 1
            inlist_x = inlist_x + [x]
            inlist_y = inlist_y + [y]
        else:
            outlist_x = outlist_x + [x]
            outlist_y = outlist_y + [y]
    if n == 1e4:
        fig, ax1 = plt.subplots()
        ax1.scatter(inlist_x, inlist_y, color = 'red')
        ax1.scatter(outlist_x, outlist_y, color = 'green')
        ax1.set_aspect('equal', adjustable='box')
        plt.text(0.86, 0.96, str((4*n_in)/n))
    print(((4*n_in)/n))
generatePi(1e4)

        
```
